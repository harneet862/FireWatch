```python
def angle_to_custom_direction_index(angle):
    # Define the directions with their custom indices
    direction_indices = {
        'SW': 0,
        'S': 1,
        'W': 2,
        'E': 3,
        'NW': 4,
        'CLM': 5,
        'N': 6,
        'SE': 7,
        'NE': 8
    }

    # Define the angle ranges for each direction
    ranges = {
        'SW': (225, 270),
        'S': (180, 225),
        'W': (270, 315),
        'E': (45, 90),
        'NW': (315, 360),
        'N': (0, 45),
        'SE': (135, 180),
        'NE': (90, 135)
    }

    # Normalize the angle to be within [0, 360)
    angle = angle % 360

    # Find the corresponding direction index
    for direction, (start, end) in ranges.items():
        if start <= angle < end or (start > end and (angle >= start or angle < end)):
            return direction_indices[direction]

    # If the angle falls exactly at 0 or 360, it should be 'N'
    if angle == 0 or angle == 360:
        return direction_indices['N']

    # Handle special case for 'CLM' if it has a specific meaning
    # For now, assuming CLM is not associated with a particular angle range
    # This part may need to be adjusted based on the context of 'CLM'

    return direction_indices['CLM']
```

```python
for i in range(24585):
  nofire_data['wind_direction'].iloc[i] = angle_to_custom_direction_index(nofire_data['wind_direction'].iloc[i])
```
